# Catalogs and Schema

```python
%sql
CREATE CATALOG IF NOT EXISTS ${catalogName}
MANAGED LOCATION 'abfss://bronze@${storageName}.dfs.core.windows.net/';
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS ${catalogName}.bronze;
CREATE SCHEMA IF NOT EXISTS ${catalogName}.silver;
CREATE SCHEMA IF NOT EXISTS ${catalogName}.gold;
CREATE SCHEMA IF NOT EXISTS ${catalogName}.raw;
```

```python
%sql
CREATE EXTERNAL VOLUME IF NOT EXISTS ${catalogName}.raw.raw_files
LOCATION 'abfss://raw@${storageName}.dfs.core.windows.net/'
COMMENT 'Volumen con los CSV originales del proyecto Supermarket (capa RAW)';
```
